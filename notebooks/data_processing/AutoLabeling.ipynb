```python
REVIEWS = "../../data/filtered/texas_reviews_filtered.csv"
LABELS = pd.read_csv("../../data/labels/texas_reviews_labels.csv")
PROCESSED = pd.read_csv("../../data/processed/texas_processed.csv")
reviews = pd.read_csv(REVIEWS)
PROCESSED['label'].describe()
```

```python
m_bad = []
m_good = []

ignore_list = ['exactly as described']
filter_list = ['not as pictured', 
               'is broken', 'was broken', 
               'are misleading', "is misleading", "was misleading","very misleading", "were misled", 'misleading description', 
               'nothing like the pictures', 
               'they are liars', 
               'not as advertised',
               'not as described',
               'not as listed',
               'we were lied to',
               'lied to us',
               'does not work', 'doesnt work', "doesn't work"]

def filtering(string):
    if any(word in string for word in filter_list) and any(word not in string for word in ignore_list):
        return True
    else:
        return False
            
for i in range(len(reviews)):
    if filtering(reviews['comments'][i]):
        if reviews['sentiment'][i] > 3:
            m_good.append(reviews['id'][i])
        elif reviews['sentiment'][i] <= 3:
            m_bad.append(reviews['id'][i])
            
print(len(m_bad))
print(len(m_good))
```

```python
reviews = pd.read_csv(REVIEWS)
reviews['auto_label'] = 'none'
for i in range(len(reviews)):
    if reviews['id'][i] in m_bad:
        reviews['auto_label'][i] = 'mbad'
    if reviews['id'][i] in m_good:
        reviews['auto_label'][i] = 'mgood'
```

```python
reviews['auto_label'].describe()
```

```python
mbad_df = reviews[reviews['auto_label'] == 'mbad'] 
mgood_df = reviews[reviews['auto_label'] == 'mgood'] 
```

```python
df = result = pd.concat([mbad_df,mgood_df])
```

```python
res = pd.merge(LABELS, df, on=['id'])
```

```python
auto_p = PROCESSED

for i in range(len(auto_p)):
    if auto_p['review_id'][i] in m_bad:
        auto_p['label'][i] = 'mbad'
    if auto_p['review_id'][i] in m_good:
        auto_p['label'][i] = 'mgood'
```
